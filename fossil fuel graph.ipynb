```python
import pandas as pd
```

```python
data1 = pd.read_csv('Black Coal.csv')
data2 = pd.read_csv('bituminous coal.csv')
data3 = pd.read_csv('heavy fuel oil.csv')
data4 = pd.read_csv('Diesel.csv')
data5 = pd.read_csv('Gas.csv')
```

```python
data2.rename(columns={'heavy fuel oil':'bituminous coal'}, inplace=True)
```

```python
data1=data1[['year','Black Coal']]
data2=data2[['year','bituminous coal']]
data3=data3[['year','heavy fuel oil']]
data4=data4[['year','Diesel']]
data5=data5[['year','Gas']]
```

```python
from matplotlib import pyplot as plt
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1,1,1)

plt.bar(data1['year'],data1['Black Coal'],label='Black Coal')
plt.bar(data2['year'],data2['bituminous coal'],label='Bituminous Coal', bottom=data1['Black Coal'])
plt.bar(data3['year'],data3['heavy fuel oil'],label='Heavy Fuel Oil',bottom=data2['bituminous coal'])
plt.bar(data4['year'],data4['Diesel'],label='Diesel', bottom=data3['heavy fuel oil'])
plt.bar(data5['year'],data5['Gas'],label='Gas', bottom=data4['Diesel'])

plt.legend(loc="best", fontsize=8)

ax1.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,10))

ax1 = plt.subplot(3,1,1)
plt.plot(data1['year'],data1['Black Coal'],label='Black Coal')
plt.plot(data2['year'],data2['bituminous coal'],label='Bituminous Coal')

ax2 = plt.subplot(3,1,2)
plt.plot(data3['year'],data3['heavy fuel oil'],label='Heavy Fuel Oil')
plt.plot(data4['year'],data4['Diesel'],label='Diesel')

ax3 = plt.subplot(3,1,3)
plt.plot(data5['year'],data5['Gas'],label='Gas')

plt.legend(loc="best", fontsize=8)

ax1.grid()

plt.show()
```
